```python
# https://pylidc.github.io/

import pydicom as pdcm
import pylidc as pl
import matplotlib.pyplot as plt
```

```python
import glob
data_filenames = glob.glob("annotated_dataset/*.jpeg")
mask_filenames = glob.glob("annotated_dataset/*.png")
print('data_filenames =', data_filenames)
print('mask_filenames =', mask_filenames)
```

```python
dcm = pdcm.dcmread("LIDC-IDRI/LIDC-IDRI-0001/01-01-2000-30178/3000566-03192/000001.dcm")
```

```python
intercept = int(dcm.RescaleIntercept)
slope = int(dcm.RescaleSlope)
mypixels = dcm.pixel_array * slope + intercept
```

```python
plt.imshow(mypixels, cmap=plt.cm.bone)
```
