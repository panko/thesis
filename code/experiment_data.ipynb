```python
import glob
#data_filenames = glob.glob("annotated_dataset/*.jpeg")
mask_filenames = glob.glob("annotated_dataset/*.png")
#print('mask_filenames =', mask_filenames)
#print('data_filenames =', data_filenames)
names = [i.split('/')[1][:-4] for i in mask_filenames]
print('names =', names, len(names))
```

```python
#dcm = pdcm.dcmread("LIDC-IDRI/LIDC-IDRI-0001/01-01-2000-30178/3000566-03192/000001.dcm")
mask = plt.imread("annotated_dataset/" + names[0] + ".png")
mask = mask[:,:,3]
mask = mask.astype(np.bool)
mask = ~mask
plt.imshow(mask, cmap=plt.cm.bone)
```

```python
lung = plt.imread("annotated_dataset/" + names[0] + ".jpeg")
plt.imshow(lung, cmap=plt.cm.bone)
```

```python
masked = np.ma.array(lung, mask=lung*mask[:,:,np.newaxis], fill_value=0) .filled()

plt.imshow(masked)
```

```python
def get_filledarea_ratio(mask):
    '''
    mask - numpy array where the nonzero values represent the actual lung
    return - float number between 0 - 1
    '''
    return np.count_nonzero(mask) / (mask.shape[0]*mask.shape[1])
```
